```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Loading cleaned dataset (from Task 1)
location = r'C:\Users\91963\Documents\DEVELOPMENT\DS\Elevate labs\Task 1- Data Cleaning & Preprocessing\cleaned_titanic.csv'
titanic = pd.read_csv(location)

# First 5 rows
print(titanic.head())
```

```python
print(titanic.describe())
```

```python
print(titanic["Sex"].value_counts())  # Male: 1, Female: 0
print(titanic["Embarked_S"].value_counts())  # Southampton passengers
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.histplot(titanic["Age"], bins=20, kde=True)
plt.title("Age Distribution")

plt.subplot(1, 3, 2)
sns.histplot(titanic["Fare"], bins=20, kde=True)
plt.title("Fare Distribution")

plt.subplot(1, 3, 3)
sns.countplot(x="Pclass", data=titanic)
plt.title("Passenger Class Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x="Pclass", y="Fare", data=titanic)
plt.title("Fare Distribution by Class")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(titanic.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.countplot(x="Sex", hue="Survived", data=titanic)
plt.title("Survival by Sex")

plt.subplot(1, 2, 2)
sns.countplot(x="Pclass", hue="Survived", data=titanic)
plt.title("Survival by Class")
plt.show()
```

```python
# Interactive histogram for Age
fig = px.histogram(titanic, x="Age", color="Survived", nbins=20, 
                   title="Age Distribution by Survival")
fig.show()

# Interactive boxplot for Fare by Class
fig = px.box(titanic, x="Pclass", y="Fare", color="Sex", 
             title="Fare Distribution by Class & Sex")
fig.show()
```
